```python
# Reverse an Array (In-Place)

'''Task:
Reverse the array in-place
❌ No extra array
❌ No slicing ([::-1])
✅ Use two pointers

left  → starts at beginning
right → starts at end

While left < right:
swap arr[left] and arr[right]
move left forward
move right backward
That’s it. No magic.
'''

arr = [1,2,3,4,5]
left = 0
right  = len(arr) - 1

while left < right:
    arr[left] , arr[right] = arr[right] , arr[left]
    left += 1
    right -= 1
print(arr)    
```

```python
# Left Rotate Array by 1 (In-Place)
'''Rules:
❌ No extra array
❌ No slicing
✅ In-place only
✅ O(n) time
'''
arr = [1,2,3,4,5]
first = arr[0]
for i in range(1,len(arr)):
    arr[i-1] = arr[i] # shifting elements by one 
arr[len(arr)-1] = first # this is the important part to remember 
print(arr)    
```
