```python
# Rotate Array by K (In-Place)
'''Task:
Rotate the array to the right by k steps'''

'''Rules:
❌ No extra array
❌ No repeated single rotations
✅ O(n) time
✅ O(1) space'''

arr = [1,2,3,4,5,6,7]
k = 4
n = len(arr)
# we have to make -> arr = [4,5,6,7,1,2,3]

'''Steps:
Reverse the entire array
Reverse the first k elements
Reverse the remaining elements'''

def reverse(l,r):
  while l<r:
    arr[l] , arr[r] = arr[r],arr[l]
    l += 1
    r -= 1
  return arr

reverse(0,n-1) 
reverse(0,k-1)
reverse(k,n-1)   
```
