```python
arr = [4, 7, 1, 8, 5]

# Print all elements using index
for i in range(len(arr)):
    print(arr[i])
```

```python
# Print them in reverse order
for i in range(len(arr)):
   print(arr[(len(arr)-1)-i])
```

```python
# Find the sum of elements
sum = 0
for i in range(len(arr)):
    sum += arr[i]
print(sum)    
```

```python
# Find the maximum element
maximum = 0
for i in range(len(arr)):
    if arr[i] > maximum:
        maximum = arr[i]
print(maximum)        
```
