```python
arr = [0, 0, 2, 0, 3, 0, 4]
'''Task:
ðŸ‘‰ Move all zeros to the end
ðŸ‘‰ Maintain order of non-zero elements'''
print(f'Given array-> {arr}')
pos = 0 
for i in range(len(arr)):
    if arr[i]!=0:
        print(f"{arr[i]} changing with {arr[pos]}")
        arr[pos]=arr[i]
        pos+=1
        print(f"arr became -> {arr} ")

print("NOW WE WANNA REMOVE REDUNDANT VALUES AND AND 0")
for i in range(pos,len(arr)):
    print(f"pos is {pos} , means only {pos} values are NON-ZERO")
    print(f"we will start loop from {i}")
    arr[i] = 0
    print(f"arr became -> {arr}")

print(f"FINAL ARRAY IS -> {arr}")    
```

```python
# ðŸ§  Problem: Remove Duplicates from a Sorted Array
arr = [1, 1, 2, 2, 2, 3, 4, 4]
'''Goal:
Keep only unique elements
Do it in-place
Maintain order
Track logical length

ðŸ”‘ Core Idea (very simple)
Because the array is sorted:
duplicates are next to each other
we only keep an element if itâ€™s different from the last kept element
'''

pos = 1 # because 1 value always unique

for i in range(1,len(arr)):
    if arr[i] != arr[pos-1]:
        arr[pos] = arr[i]
        pos += 1

arr_1 = []
for i in range(pos):
    arr_1.append(arr[i])

print("Before deletion ", arr)
print("after deletion " , arr_1)    
```
